## PART A: BUILDING OF A LINEAR REGRESSION MODEL

### First we would love to build a Linear Regression Model Exploring the following relationships:
1. Explore if being more experienced by having more Number of years in Experience at work affects Salaries in the Tech Job Market.


- Start by defining the dependent variable(salary) and the independent variable (Experience in Years)

```python
##First we define the outcome variable(y) and the independent variable (x)
x =jobs_df1[['experience_years']]
y =jobs_df1[['salary']]
ols_data = jobs_df1[['experience_years', 'salary']]
ols_data
```

- Then formulate the Linear Regression Model

```python
ols_formula = 'salary ~ experience_years'
```

```python
## Import the OLS module from STATS.MODELS. The OLS Model is used for Fitting
from statsmodels.formula.api import ols
```

- Then fit the Regression Model

```python
ols_model = ols(formula=ols_formula, data=ols_data)
model = ols_model.fit()
```

- Obtain Model Summary Statistics

```python
model.summary()
```

### Results Findings and Interpretation of the Linear Regression Model Based on Salary ~ Experience In Years
- COEFFICIENTS_INTERPRETATIONS
1. For the experience_years variable the coefficient is 2,701. Implies that for every additional one experience in years, the Salary increases by a value of 2,701.

### Explaining the Uncertainity of our Regression Model Predictive Power.
- From the Model Summary Statistics, we have a few evaluation metrics which can help us determine if the Regression Model has any Predictive Power.
- These include the R- Squared Value, Confidence Intervals and The P-Value.

R-Squared Value Interpretations: For a value of 0.192 translating to an approximate value of 19.2% that explains the variation of the Salary explained by the Variable Experience in Years.However a 19.2% value holds little power in the predictive model ability to use Experience in years to predict the amount of Salary received.

The confidence Interval [2679.41,2722.92] implies the chance of producing the true Beta value (Experience in Years) of 2701.

The P-Value is used to determine whether our COEFFICIENT is STATISTICALY SIGNIFICANT or not! Having that the P-Value is less than value of 0.05, we ccan be pretty confident that the Coeffficient value of 2,701 is Statistically Significant.

### Checking for Model Assumption
A Linear Regression Models have Assumptions to be Checked to determine if or not the Model is capable of predictive power.
These Include:
- LINEARITY: The INDEPENDENT Variable is Linearly related to the INDEPENDENT Variable
- NORMALITY: The error terms is normally distributed.
- HOMOSCEDACITY: The variance of the errors is constant across the model.
- INDEPENDENCE: Each and every observation is independent.

A. Check for NORMALITY Assumptions.
- Using a Histogram Plot

```python
#First We Subset the X variable from the data
x = ols_data['experience_years']
```

```python
# We then Obtain the predictions from the model
fitted_values = model.predict(x)
```

```python
# Then Obtain the residuals
residuals = model.resid
```

```python
# Use a histogram to check for the Normality assumption
fig = sns.histplot(residuals)
fig.set_xlabel('Residual Value')
fig.set_title('Histogrm of Residuals')
plt.show()
```

- Using a Quantile - Quantile (Q-Q Plot)

```python
#We then Use the Quantiles-Quantiles (Q-Qplot) to check for the Normality assumptions
fig = sm.qqplot(residuals, line='s')
plt.title('Q-Q Plot for Residuals')
plt.show()
```

#### NORMALITY Assumption Interpretations
The Normality Assumption has been fulfilled in both plots. Case exmple the Histogram shows a Normal Distribution whereas the Q-Q Plot show a linear increase across the data points.

- B. Check for Homoscedacity Assumption

```python
# We use a Scatterplot
fig = sns.scatterplot(x=fitted_values, y=residuals)
fig.axhline(0)
fig.set_xlabel("Fitted Values")
fig.set_ylabel("Residuals")
plt.show()
```

## PART B: BUILDING A MULTIPLE LINEAR REGRESSION MODEL
In this part we would like to analyze if a combination of:
- Experience in Years
- Skills Count
- Certifications Number.
affects the amount of salary a Tech employee earns.

```python
jobs_df1.head(2)
```

```python
# For this case we assume that prediction of the Jobs Salary based on More than one Continous variable(experience_years, skills and certications
## STEP 1. Create a Subset for the X and Y variables
jobs_x = jobs_df1[['experience_years', 'skills_count', 'certifications']]
jobs_y = jobs_df1[['salary']]
```

```python
## STEP 2. IMPORT The train_split function from SCI-KIT LEARN
from sklearn.model_selection import train_test_split
```

```python
## STEP 3. Create training dataset nd a Holdout sample datasets
x_train, x_test, y_train, y_test = train_test_split(jobs_x, jobs_y, test_size=0.3, random_state=42)
    #The test size=0.3 (30%) tells the function the PROPORTION of data to be in the Hold Out Sample. We hold 30% of the data to test the model
    #The random_state=42 is for reproducibility purposes
```

```python
## STEP 4. MODEL CONSTRUCTION
ols_formula = 'salary ~ experience_years + skills_count + certifications'
    #We define the OLS formula for the Single Y variable and the X Variables
```

```python
# We import the ols() function
from statsmodels.formula.api import ols
```

```python
# We create the OLS Data Frame
ols_data = pd.concat([x_train, y_train], axis=1)
```

```python
# We then create an OLS Obect and fit the MODEL
OLS = ols(formula=ols_formula, data=ols_data)
model = OLS.fit()
```

```python
## Obtain the Summary result of the Model AS:
model.summary()
```

### Model Bulding Again after Adding More Variables
- After defining more variables into our data, we select the variable features which are to be Icluded in the model building

```python
features = ['experience_years','skills_count','certifications','educlev_encoding', 'compsize_encoding','expeience_years_squared']
```

## Divide the data into Train_Test_Split

```python
x_train, x_test, y_train, y_test = train_test_split(x_variable, y_variable, test_size=0.3, random_state=42)
```

### Obtain the Scale Features.
- First we import the Standard Scaler Library Package

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

### We Fit The Model
- We also import the Linear Regression Library from sklearn.linear_model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train_scaled, y_train)
```

```python
### Checking for any improvements from the previous model?
r2_train = model.score(x_train_scaled, y_train)
r2_test = model.score(x_test_scaled, y_test)

print(f"Training R²: {r2_train:.3f}")
print(f"Test R²: {r2_test:.3f}")
```

```python
coefficients = pd.DataFrame({
    'Feature': features,
    'Coefficient': model.coef_
})
print("\nFeature Importance:")
print(coefficients.sort_values('Coefficient', ascending=False))
```
